```python
import numpy as np         
import scipy.stats as stats 

data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
data_length = len(data)

mean_data = np.mean(data)
print(mean_data)

s = np.std(data, ddof=1)
print(s)

standard_error = s / np.sqrt(data_length)

t_dist = stats.t.interval(confidence=0.99, 
                        df=data_length-1, 
                        loc=mean_data, 
                        scale=standard_error)

print(f"99% sure interval using t dist: {t_dist}")

sigma = 2.0

z_dist = stats.norm.interval(confidence=0.99, 
                           loc=mean_data, 
                           scale=sigma / np.sqrt(data_length))

print(f"99% sure (with known sigma) interval using z dist: {z_dist}")
```
